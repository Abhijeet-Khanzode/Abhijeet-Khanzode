# **`FINAL OUTPUT WITH CODE AND ALGORITHM `**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split
from collections import defaultdict

# Function to read CSV data from URL
def read_csv_from_url(url):
    return pd.read_csv(url)

# Function to preprocess the data and build the recommendation model
def preprocess_and_build_model(df):
    # Check if 'Movie_Title' column exists
    if 'Movie_Title' not in df.columns:
        raise ValueError("The dataset doesn't contain a 'Movie_Title' column.")

    # Vectorize the movie titles
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(df['Movie_Title'])

    # Convert the dataset into the surprise format
    reader = Reader(rating_scale=(1, 5))
    data = Dataset.load_from_df(df[['Movie_ID', 'Movie_ID', 'Movie_Vote']], reader)

    # Split the dataset into training and testing sets
    trainset, testset = train_test_split(data, test_size=0.25)

    # Use the KNNBasic algorithm for item-based collaborative filtering
    algo = KNNBasic(sim_options={'user_based': False})
    algo.fit(trainset)

    return tfidf_matrix, algo

# Function to find the closest match
def find_closest_match(title, tfidf_matrix, titles):
    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
    idx = titles[titles.str.lower() == title.lower()].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    return [i[0] for i in sim_scores[1:2]]

# Function to get top 30 movie recommendations
def get_top_n_recommendations(algo, userId, df, n=30):
    # Get the list of all movie IDs
    movie_ids = df['Movie_ID'].unique()

    # Predict ratings for all movies for the given user
    predictions = [algo.predict(userId, movie_id) for movie_id in movie_ids]

    # Sort predictions by estimated rating
    top_predictions = sorted(predictions, key=lambda x: x.est, reverse=True)[:n]

    # Get the movie titles for the top predictions
    top_movie_ids = [pred.iid for pred in top_predictions]
    top_movies = df[df['Movie_ID'].isin(top_movie_ids)]['Movie_Title']

    return top_movies

# Main function
def main():
    # Read CSV data from URL
    url = "https://github.com/YBI-Foundation/Dataset/raw/main/Movies%20Recommendation.csv"
    df = read_csv_from_url(url)

    # Preprocess data and build recommendation model
    tfidf_matrix, algo = preprocess_and_build_model(df)

    # Get user input
    user_input = input("Enter a movie name: ")

    # Find closest match
    closest_idx = find_closest_match(user_input, tfidf_matrix, df['Movie_Title'])
    closest_match = df.iloc[closest_idx]['Movie_Title'].values[0]
    print(f"Closest match to '{user_input}' is '{closest_match}'")

    # Assume userId = 1 for demonstration
    user_id = 1

    # Get top 30 movie recommendations for the user
    recommended_movies = get_top_n_recommendations(algo, user_id, df, n=30)

    print("Recommended movies:")
    print(recommended_movies)

if __name__ == "__main__":
    main()
```
